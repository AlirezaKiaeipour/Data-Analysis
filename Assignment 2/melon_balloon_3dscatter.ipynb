```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Melon width = 20
# Melon length = 30
# Melon weight = 4
# Balloons width = 30
# Balloons length = 30
# Balloons weight = 0.01
```

```python
N = 1000
std = 2
std_melon_weight = 0.5
std_balloons_weight = 0.001
```

```python
melon_width = np.random.normal(20,std,N)
melon_length = np.random.normal(30,std,N)
melon_weight = np.random.normal(4,std_melon_weight,N)

balloon_width = np.random.normal(30,std,N)
balloon_length = np.random.normal(30,std,N)
balloon_weight = np.random.normal(0.01,std_balloons_weight,N)
```

```python
width = np.concatenate((melon_width,balloon_width))
length = np.concatenate((melon_length,balloon_length))
weight = np.concatenate((melon_weight,balloon_weight))
```

```python
x = np.array([width,length,weight]).T

melon_label = np.zeros(N,np.int8)
balloon_label = np.ones(N,np.int8)
y = np.concatenate((melon_label,balloon_label))
```

```python
## 2D Cartesian space

plt.scatter(x[:N,0],x[:N,1],x[:N,2],c="yellow")
plt.scatter(x[N:,0],x[:N,1],x[:N,2],c = "red")
plt.xlabel("width")
plt.ylabel("length")
```

```python
fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(projection="3d")
ax.scatter(x[:N,0],x[:N,1],x[:N,2],c="yellow")
ax.scatter(x[N:,0],x[:N,1],x[:N,2],c = "red")
ax.set_xlabel("width")
ax.set_ylabel("length")
ax.set_zlabel("weight")
```
