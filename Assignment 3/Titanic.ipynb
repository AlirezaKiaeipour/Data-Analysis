```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("titanic.csv")
data.head()
```

```python
male = data[data["Sex"]=="male"].reset_index()
female = data[data["Sex"]=="female"].reset_index()
```

```python
plt.figure(figsize=(10,5))
plt.hist(male["Age"],bins=20,label="Male")
plt.hist(female["Age"],bins=20,alpha=0.8,label="Female")
plt.legend()
plt.xlabel("Age")
plt.ylabel("Frequently")
plt.title("Age histogram chart")
```

```python
sns.histplot(male["Age"],color="Blue",kde="True")
sns.histplot(female["Age"],color="orange",kde="True")
p = sns.lineplot()
sns.set(rc={'figure.figsize':(8,5)})
p.set_xlabel("Age", fontsize = 15)
p.set_ylabel("Frequently", fontsize = 15)
p.set_title("Age histogram chart",fontsize = 15)
p.legend(labels = ["Male","Female"])
```

```python
pclass1 = data[data["Pclass"]==1].reset_index()
pclass2 = data[data["Pclass"]==2].reset_index()
pclass3 = data[data["Pclass"]==3].reset_index()
```

```python
plt.figure(figsize=(10,5))
plt.hist(pclass1["Fare"],bins=10,label="Pclass 1")
plt.hist(pclass2["Fare"],bins=10,alpha=0.8,label="Pclass 2")
plt.hist(pclass3["Fare"],bins=10,alpha=0.8,label="Pclass 3")
plt.legend()
plt.xlabel("Fare")
plt.ylabel("Frequently")
plt.title("Fare histogram chart")
```

```python
sns.histplot(pclass1["Fare"],color="Blue",bins=10)
sns.histplot(pclass2["Fare"],color="orange",bins=10)
sns.histplot(pclass3["Fare"],color="green",bins=10)
p = sns.lineplot()
sns.set(rc={'figure.figsize':(8,5)})
p.set_xlabel("Fare", fontsize = 15)
p.set_ylabel("Frequently", fontsize = 15)
p.set_title("Fare histogram chart",fontsize = 15)
p.legend(labels = ["Pclass 1","Pclass 2","Pclass 3"])
```
